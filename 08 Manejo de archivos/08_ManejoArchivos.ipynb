```python


# ----------------------------------------------
# Gestión de productos con archivos de texto
# ----------------------------------------------

# Archivo de texto
ARCHIVO = "productos.txt"

# 1️⃣ Crear archivo inicial con 3 productos
def crear_archivo_inicial():
    try:
        open(ARCHIVO, "x").write("Lapicera,120.5,30\nCuaderno,450.0,50\nRegla,80.0,100\n")
        print("Archivo creado con productos iniciales.")
    except FileExistsError:
        pass  # Si ya existe, no lo crea de nuevo

# 2️⃣ Leer productos y cargarlos en una lista de diccionarios
def leer_productos():
    productos = []
    with open(ARCHIVO, "r") as f:
        for linea in f:
            nombre, precio, cantidad = linea.strip().split(",")
            productos.append({
                "nombre": nombre,
                "precio": float(precio),
                "cantidad": int(cantidad)
            })
    return productos

# 3️⃣ Mostrar productos
def mostrar_productos(productos):
    for p in productos:
        print(f"Producto: {p['nombre']} | Precio: ${p['precio']} | Cantidad: {p['cantidad']}")

# 4️⃣ Agregar producto desde teclado
def agregar_producto():
    nombre = input("Nombre: ")
    precio = input("Precio: ")
    cantidad = input("Cantidad: ")
    with open(ARCHIVO, "a") as f:
        f.write(f"{nombre},{precio},{cantidad}\n")
    print("Producto agregado correctamente.")

# 5️⃣ Buscar producto por nombre
def buscar_producto():
    nombre_buscar = input("Ingrese el nombre del producto: ").lower()
    productos = leer_productos()
    encontrado = False
    for p in productos:
        if p["nombre"].lower() == nombre_buscar:
            print(f"Producto encontrado -> Nombre: {p['nombre']} | Precio: ${p['precio']} | Cantidad: {p['cantidad']}")
            encontrado = True
    if not encontrado:
        print("Producto no encontrado.")

# 6️⃣ Guardar lista actualizada (sobrescribir archivo)
def guardar_productos(productos):
    with open(ARCHIVO, "w") as f:
        for p in productos:
            f.write(f"{p['nombre']},{p['precio']},{p['cantidad']}\n")
    print("Archivo guardado correctamente.")

# ----------------------------------------------
# Menú principal con match / case
# ----------------------------------------------
def menu():
    crear_archivo_inicial()
    while True:
        print("\n--- MENÚ DE PRODUCTOS ---")
        print("1. Mostrar productos")
        print("2. Agregar producto")
        print("3. Buscar producto por nombre")
        print("4. Modificar cantidad de un producto")
        print("5. Salir")

        opcion = input("Elija una opción: ")

        match opcion:
            case "1":
                productos = leer_productos()
                mostrar_productos(productos)
            case "2":
                agregar_producto()
            case "3":
                buscar_producto()
            case "4":
                productos = leer_productos()
                nombre = input("Ingrese el nombre del producto a modificar: ").lower()
                for p in productos:
                    if p["nombre"].lower() == nombre:
                        nueva_cantidad = input("Nueva cantidad: ")
                        p["cantidad"] = int(nueva_cantidad)
                guardar_productos(productos)
            case "5":
                print("Saliendo del programa...")
                break
            case _:
                print("Opción inválida. Intente de nuevo.")

# Ejecutar el programa
menu()
```
